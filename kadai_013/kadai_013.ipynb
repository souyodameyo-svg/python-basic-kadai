```python
def calculate_tax_included(price: int, tax_rate: int) -> int:
    # 消費税額を計算し、元の金額に加える
    # tax_rateが10の場合、10 / 100 = 0.1 を掛ける
    total = price + (price * tax_rate / 100)
    return int(total)  # 小数点以下を切り捨てて整数で返す

# --- 実行と結果の確認 ---

# 第1引数に金額「110」、第2引数に税率「10」を設定
result = calculate_tax_included(110, 10)

print(f"税込金額は {result} 円です")
```
